## ðŸ“¥ Load Dataset

We load the IMDB dataset into a DataFrame called movies to begin exploration.

```python
import pandas as pd

movies = pd.read_csv(r'C:\Users\User\OneDrive\Desktop\Data Analyst Builder\Projects\Movie Genres Data Analysis Project\Data\imdb_movies.csv')
movies.head()
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option('display.max.columns',22)
```

```python
movies.head()
```

```python
#Remove Duplicate rows
movies.drop_duplicates(inplace = True)
```

```python
#Remove all null values from the genres column
movies.dropna(subset = ['genres'], inplace = True)
```

```python
#New column to calculate profit
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
#Cleaned data with removed irrelevant columns
movies_genre = movies[['popularity', 'budget', 'revenue', 'profit', 'original_title', 'runtime', 'genres','release_date','vote_count','vote_average']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

## Q1: Which genres are the most common?

We group the data by genre and count the number of movies in each to determine the most frequently produced genres.

```python
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending = True)
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %', autopct = '%1.1f%%', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', color = 'DarkBlue', figsize = (10,9))
```

ðŸŽ¬ **Observation:** The most common genres are Drama, Comedy, Thriller and Action. This indicates a preference among studios for producing these genres.
